```python
# a) Extraer la información del archvivo
import pandas as pd

# Ruta del archivo en Google Drive
file_path = '/content/drive/My Drive/Pandas y Numpy Reto/country_vaccinations.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path)

# Mostrar las primeras filas del dataframe
df.head()
```

```python
# b) Mostrar la estructura y tipos de datos de cada columna
df.info()

```

```python
# c) Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria)
vacunas_por_compania = df.groupby('vaccines')['total_vaccinations'].sum()

# Mostrar el resultado
vacunas_por_compania
```

```python
# d) Sumar la cantidad total de vacunas aplicadas en todo el mundo
vacunas_totales = df['total_vaccinations'].sum()

# Mostrar el resultado
vacunas_totales
```

```python
# e) Calcular el promedio de vacunas aplicadas por país
promedio_vacunas_por_pais = df.groupby('country')['total_vaccinations'].mean()

# Mostrar el resultado
promedio_vacunas_por_pais
```

```python
# f) Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo
vacunas_29_01_21 = df[df['date'] == '2021-01-29']['total_vaccinations'].sum()

# Mostrar el resultado
vacunas_29_01_21
```

```python
# g) Crear la columna 'diferencias' con la diferencia entre 'daily_vaccinations' y 'daily_vaccinations_raw'
df['diferencias'] = df['daily_vaccinations'] - df['daily_vaccinations_raw']

# Crear el nuevo dataframe conDiferencias
conDiferencias = df.copy()

# Mostrar las primeras filas del nuevo dataframe
conDiferencias.head()
```

```python
# h) Obtener la fecha más reciente y la más antigua
fecha_mas_reciente = df['date'].max()
fecha_mas_antigua = df['date'].min()

# Calcular el periodo de tiempo
periodo_tiempo = fecha_mas_reciente - fecha_mas_antigua

# Mostrar el resultado
periodo_tiempo
```

```python
# i) Crear la columna 'canVac' con la cantidad de vacunas utilizadas cada día
df['canVac'] = df['vaccines'].str.split(',').apply(len)

# Crear el nuevo dataframe conCantidad
conCantidad = df.copy()

# Mostrar las primeras filas del nuevo dataframe
conCantidad.head()
```

```python
# j) Filtrar los registros antes del 20 de diciembre de 2020
antes20 = df[df['date'] < '2020-12-20']

# Mostrar las primeras filas del nuevo dataframe
antes20.head()
```

```python
# k) Filtrar los registros donde se haya utilizado la vacuna Pfizer
pfizer = df[df['vaccines'].str.contains('Pfizer', na=False)]

# Mostrar las primeras filas del nuevo dataframe
pfizer.head()
```

```python
# l) Almacenar los dataframes en un archivo de Excel con diferentes hojas
with pd.ExcelWriter('/content/drive/My Drive/Pandas y Numpy Reto/resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
```
